```python
def graph(n, nodes):
    g = {}
    for i in nodes:
        g[i] = set()
    return g

def addNode(g, new):
    g[new] = set()

def addEdge(g, node1, node2):
    g[node1].add(node2)
    g[node2].add(node1)

def BFS(g, start, goal):
    q = [start]
    path = []
    visited = [start]
    while q:
        val = q.pop(0)
        path.append(val)
        if val == goal:
            return path
        for i in g[val]:
            if i not in visited:
                q.append(i)
                print(q)
                visited.append(i)

def DFS(g, start, goal):
    s = [start]
    visited = [start]
    path =[]
    while s:
        val = s.pop()
        path.append(val)
        if val == goal:
            return path
        for i in g[val]:
            if i not in visited:
                s.append(i)
                visited.append(i)
    
```

```python
while True:
    ch = int(input("1.Create Graph\n2.Add a node\n3.Add an edge\n4.Print Graph\n5.Print AdjacencyList\n6.BFS\n7.DFS\n10.exit\nEnter your choice:"))
    if(ch == 1):
        n = int(input("Enter no. of nodes:"))
        nodes = list(input("Enter the nodes:").split(" "))
        g = graph(n, nodes)
    elif(ch == 2):
        new = input("Enter new node:")
        if new not in nodes:
            nodes.append(new)
            addNode(g, new)
            print(">>>Node added!")
        else:
            print(">>>Node already exists!")
    elif(ch == 3):
        n= int(input("Enter no. of edges:"))
        for i in range(n):
            node1 = input("Node 1:")
            node2 = input("Node 2:")
            if node1 in nodes and node2 in nodes:
                addEdge(g,node1,node2)
                print(">>>Edge added!")
            else:
                print(">>>Invalid Nodes!")
    elif(ch == 4):
        print(g)
    elif(ch == 5):
        node = input("Enter node:")
        if node in nodes:
            print(node," : ",g[node])
        else:
            print("Invalid Node!")
    elif(ch == 6):
        start = input("Enter Start node:")
        goal = input("Enter goal node:")
        path = BFS(g, start, goal)
        print(path)
    elif(ch == 7):
        start = input("Enter Start node:")
        goal = input("Enter goal node:")
        path = DFS(g, start, goal)
        print(path)
    else:
        break
```
