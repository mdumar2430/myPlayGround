```python
def initGraph(n):
    graph = {}
    for i in range(n):
        key = i + 1
        graph[key] = {}
    return graph

def add_node(graph, newNode):
    if newNode not in graph:
        graph[newNode] = {}
        return 1
    return 0

def add_edge(graph, node1, node2, cost):
    if node1 in graph and node2 in graph:
        graph[node1][node2] = cost
        graph[node2][node1] = cost
        return 1
    return 0

def print_graph(graph):
        if(graph == {}):
            return 0
        print(graph)
        return 1

def print_adjList(graph, node):
    if node in graph:
        print(node, graph[node])
        return 1
    return 0

def DFS(graph,root,end):
    visited = []
    queue = []
    queue.append(root)
    while queue != []:
        root = queue.pop()
        if not root in visited:
            visited.append(root)
            if root == end:
                break;
            for n in graph[root] :
                queue.append(n)
    return visited
        
def BFS(graph,root,end):
    visited = []
    queue = []
    queue.append(root)
    while queue != []:
        root = queue.pop(0)
        if not root in visited:
            visited.append(root)
            if root == end:
                break;
            for n in graph[root] :
                queue.append(n)
    return visited

while True:
    print("\n1. Initialize Graph\n2. Add Edge\n3. Add Node\n4. Print Graph\n5. Print Adjacency List\n6. DFS\n7. BFS\n8. Exit")
    ch = int(input())
    if(ch == 1):
        n = int(input("Enter no. of Nodes:"))
        g = initGraph(n)
    elif(ch == 2):
        n1 = int(input("Start:"))
        n2 = int(input("End:"))
        cost = int(input("Cost:"))
        if(add_edge(g, n1, n2, cost)):
            print("-->Edge Added Successfully!")
        else:
            print("-->Invalid edge, Try Again!")
    elif(ch == 3):
        new = int(input("Enter new node:"))
        if(add_node(g, new)):
            print("-->Node Added Succesfully!")
        else:
            print("-->Node already available, Try Again!")
    elif(ch == 4):
        if(not print_graph(g)):
            print("No graph!")
    elif(ch == 5):
        node = int(input("Enter node:"))
        if(not print_adjList(g, node)):
            print("-->Enter Valid Node!")
    elif(ch == 6):
        start = int(input("Start:"))
        end = int(input("Goal:"))
        path = DFS(g, start, end)
        print(path)
    elif(ch == 7):
        start = int(input("Start:"))
        end = int(input("Goal:"))
        path = BFS(g, start, end)
        print(path)
    else:
        g={}
        break
```
