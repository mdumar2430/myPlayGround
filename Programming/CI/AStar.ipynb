```python
def graph(n, nodes):
    g = {}
    for i in nodes:
        g[i] = set()
    return g

def addNode(g, new):
    g[new] = set()

def addEdge(g, node1, node2):
    g[node1].add(node2)
    g[node2].add(node1)

def addHcost(nodes, h):
    temp = dict(zip(nodes, h))
    return temp

def a_star(g, start, goal, hcost):
    f_list =[[hcost[start],[start]]]
    visited_edge = []
    while True:
        val = f_list.pop(0)
        node = val[1][-1]
        if node == goal:
            return (val[1])           
        for i in g[node]:
            if (node, i) not in visited_edge and (i, node) not in visited_edge:
                temp = val[1].copy()
                temp.append(i)
                cost = hcost[i]
                f_list.append([cost,temp])
                visited_edge.append((node, i))
        f_list = sorted(f_list)
        print(f_list)
```

```python
while True:
    print("\n1. Initialize Graph\n2. Add Edge\n3. Add Node\n4. Print Graph\n5. Add Hcost\n6. Astar Serach\n8. Exit")
    ch = int(input())
    if(ch == 1):
        n = int(input("Enter no. of nodes:"))
        nodes = list(input("Enter the nodes:").split(" "))
        g = graph(n, nodes)
    elif(ch == 3):
        new = input("Enter new node:")
        if new not in nodes:
            nodes.append(new)
            addNode(g, new)
            print(">>>Node added!")
        else:
            print(">>>Node already exists!")
    elif(ch == 2):
        n= int(input("Enter no. of edges:"))
        for i in range(n):
            node1 = input("Node 1:")
            node2 = input("Node 2:")
            if node1 in nodes and node2 in nodes:
                addEdge(g,node1,node2)
                print(">>>Edge added!")
            else:
                print(">>>Invalid Nodes!")
    elif(ch == 4):
        print(g)
    elif(ch == 5):
        hcost = []
        for i in nodes:
            print(i,":")
            val = input()
            hcost.append(val)
        hc = addHcost(nodes, hcost)
    elif(ch == 6):
        start = input("Start node:")
        goal = input("Goal node:")
        path = a_star(g, start, goal, hc)
        print(path)  
    else:
        break
    
```
