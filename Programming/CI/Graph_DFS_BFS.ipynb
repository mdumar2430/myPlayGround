```python
def initGraph(n):
    graph = {}
    for i in range(n):
        key = i + 1
        graph[key] = set([])
    return graph

def add_node(graph, newNode):
    if newNode not in graph:
        graph[newNode] = set([])
        return 1
    return 0

def add_edge(graph, node1, node2):
    if node1 in graph and node2 in graph:
        graph[node1].add(node2)
        graph[node2].add(node1)
        return 1
    return 0

def print_graph(graph):
        if(graph == {}):
            return 0
        print(graph)
        return 1

def print_adjList(graph, node):
    if node in graph:
        print(node, graph[node])
        return 1
    return 0

def DFS(graph, start, visited):
    stack = []
    path = []
    visited.append(start)
    stack.append(start)
    while stack:
        val = stack.pop()
        path.append(val)
        for node in graph[val]:
            if node not in visited:
                visited.append(node)
                stack.append(node)
    return path
        
def BFS(graph, start, visited):
    queue =[]
    path=[]
    visited.append(start)
    queue.append(start)
    while queue:
        val = queue.pop(0)
        path.append(val)
        for node in graph[val]:
            if node not in visited:
                visited.append(node)
                queue.append(node)
    return path

while True:
    print("\n1. Initialize Graph\n2. Add Edge\n3. Add Node\n4. Print Graph\n5. Print Adjacency List\n6. DFS\n7. BFS\n8. Exit")
    ch = int(input())
    if(ch == 1):
        n = int(input("Enter no. of Nodes:"))
        g = initGraph(n)
    elif(ch == 2):
        n1 = int(input("Start:"))
        n2 = int(input("End:"))
        if(add_edge(g, n1, n2)):
            print("-->Edge Added Successfully!")
        else:
            print("-->Invalid edge, Try Again!")
    elif(ch == 3):
        new = int(input("Enter new node:"))
        if(add_node(g, new)):
            print("-->Node Added Succesfully!")
        else:
            print("-->Node already available, Try Again!")
    elif(ch == 4):
        if(not print_graph(g)):
            print("No graph!")
    elif(ch == 5):
        node = int(input("Enter node:"))
        if(not print_adjList(g, node)):
            print("-->Enter Valid Node!")
    elif(ch == 6):
        visited = []
        start = int(input("Root:"))
        path = DFS(g, start, visited)
        print(path)
    elif(ch == 7):
        visited = []
        start = int(input("Root:"))
        path = BFS(g, start, visited)
        print(path)
    else:
        g={}
        break
```
